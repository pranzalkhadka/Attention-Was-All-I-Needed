```python
import torch
import math
from torch import nn
import torch.nn.functional as F
```

```python
def scaled_dot_product(q, k, v, mask=None):
    # This is later used to scale the dot product of QUERY and KEY vectors
    # 30, 8, 200, 64 --> This will be the shape of input to the function
    d_k = q.size()[-1] # 64
    # The key vector is transposed such that the sequence length and size of each word is interchanged keeping batch size as it is
    scaled = torch.matmul(q, k.transpose(-1, -2)) / math.sqrt(d_k) # 30, 8, 200, 200
    print(f"scaled.size() : {scaled.size()}")
    # In decoder section, we use this mask to ensure that the attention is applied only to the current and previous positions. This is not used in the encoder
    if mask is not None:
        print(f"Adding mask of shape {mask.size()} ") # Here pytorch automatically adds 200 * 200 for every batch and every head
        scaled += mask
    # Passing the output through softmax activation function
    attention = F.softmax(scaled, dim=-1) # 30, 8, 200, 200
    # Performing dot product between attention and VALUE vectors
    values = torch.matmul(attention, v) # 30, 8, 200, 64
    # These values now carry more and better context than the original representation
    return values, attention
```

```python
class MultiHeadAttention(nn.Module):

    def __init__(self, d_model, num_heads):
        super().__init__()
        # Vector dimension of each word
        self.d_model = d_model # 512
        # Number of heads in the multi head attention
        self.num_heads = num_heads # 8
        # Vector dimension of each head
        self.head_dim = d_model // num_heads # 64
        # Generating the query, key and value vectors for all heads
        self.qkv_layer = nn.Linear(d_model , 3 * d_model) # 512 * 1536 (512 * 3)
        self.linear_layer = nn.Linear(d_model, d_model) # 512 * 512
    
    def forward(self, x, mask=None):
        batch_size, max_sequence_length, d_model = x.size() # 30, 200, 512
        print(f"x.size(): {x.size()}")
        qkv = self.qkv_layer(x) # 30, 200, 1536 (512 * 3)
        print(f"qkv.size(): {qkv.size()}")
        qkv = qkv.reshape(batch_size, max_sequence_length, self.num_heads, 3 * self.head_dim) # 30, 200, 8, 192 (64 * 3)
        print(f"qkv.size(): {qkv.size()}")
        qkv = qkv.permute(0, 2, 1, 3) # 30, 8, 200, 192
        print(f"qkv.size(): {qkv.size()}")
        q, k, v = qkv.chunk(3, dim=-1) # 30, 8, 200, 64 (192 / 3)
        print(f"q size: {q.size()}, k size: {k.size()}, v size: {v.size()}, ")
        values, attention = scaled_dot_product(q, k, v, mask) # 30, 8, 200, 64 and 30, 8, 200, 200
        print(f"values.size(): {values.size()}, attention.size:{ attention.size()} ")
        values = values.reshape(batch_size, max_sequence_length, self.num_heads * self.head_dim) # 30, 200, 512 (64 * 8)
        print(f"values.size(): {values.size()}")
        out = self.linear_layer(values)
        print(f"out.size(): {out.size()}")
        return out
```

```python
class LayerNormalization(nn.Module):
    def __init__(self, parameters_shape, eps=1e-5):
        super().__init__()
        self.parameters_shape = parameters_shape
        self.eps = eps # To prevent division by zero
        self.gamma = nn.Parameter(torch.ones(parameters_shape)) # 512 dimensional learnable parameter
        self.beta =  nn.Parameter(torch.zeros(parameters_shape)) # 512 dimensional learnable parameter

    def forward(self, inputs):
        dims = [-(i + 1) for i in range(len(self.parameters_shape))]
        # Calculating mean of the input tensor
        mean = inputs.mean(dim=dims, keepdim=True)
        print(f"Mean ({mean.size()})")
        var = ((inputs - mean) ** 2).mean(dim=dims, keepdim=True)
        # Calculating standard deviation of the input tensor
        std = (var + self.eps).sqrt()
        print(f"Standard Deviation  ({std.size()})")
        # Normalizing the input
        y = (inputs - mean) / std
        print(f"y: {y.size()}")
        # Adding the learnable parameters
        out = self.gamma * y  + self.beta
        print(f"self.gamma: {self.gamma.size()}, self.beta: {self.beta.size()}")
        print(f"out: {out.size()}")
        return out
```

```python
class PositionwiseFeedForward(nn.Module):

    def __init__(self, d_model, hidden, drop_prob=0.1):
        super(PositionwiseFeedForward, self).__init__()
        self.linear1 = nn.Linear(d_model, hidden)
        self.linear2 = nn.Linear(hidden, d_model)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(p=drop_prob)

    def forward(self, x):
        x = self.linear1(x)
        print(f"x after first linear layer: {x.size()}")
        x = self.relu(x)
        print(f"x after activation: {x.size()}")
        x = self.dropout(x)
        print(f"x after dropout: {x.size()}")
        x = self.linear2(x)
        print(f"x after 2nd linear layer: {x.size()}")
        return x
```

```python
class EncoderLayer(nn.Module):

    def __init__(self, d_model, ffn_hidden, num_heads, drop_prob):
        super(EncoderLayer, self).__init__()
        # Encoder architecture step by step
        self.attention = MultiHeadAttention(d_model=d_model, num_heads=num_heads)
        self.norm1 = LayerNormalization(parameters_shape=[d_model])
        self.dropout1 = nn.Dropout(p=drop_prob)
        self.ffn = PositionwiseFeedForward(d_model=d_model, hidden=ffn_hidden, drop_prob=drop_prob)
        self.norm2 = LayerNormalization(parameters_shape=[d_model])
        self.dropout2 = nn.Dropout(p=drop_prob)

    def forward(self, x):
        residual_x = x # 30, 200, 512
        print("------- ATTENTION 1 ------")
        x = self.attention(x, mask=None) # 30, 200, 512
        print("------- DROPOUT 1 ------")
        x = self.dropout1(x) # 30, 200, 512
        print("------- ADD AND LAYER NORMALIZATION 1 ------")
        x = self.norm1(x + residual_x)
        residual_x = x
        print("------- ATTENTION 2 ------")
        x = self.ffn(x)
        print("------- DROPOUT 2 ------")
        x = self.dropout2(x)
        print("------- ADD AND LAYER NORMALIZATION 2 ------")
        x = self.norm2(x + residual_x)
        return x
```

```python
class Encoder(nn.Module):
    def __init__(self, d_model, ffn_hidden, num_heads, drop_prob, num_layers):
        super().__init__()
        self.layers = nn.Sequential(*[EncoderLayer(d_model, ffn_hidden, num_heads, drop_prob)
                                     for _ in range(num_layers)])

    def forward(self, x):
        x = self.layers(x)
        return x
```

```python
# This is the dimension of each word throughout the encoder
d_model = 512

# This is the number of heads in multi head attention. So, 8 different attention blocks will run in parallel
num_heads = 8

# This is the dropiout rate of neurons
drop_prob = 0.1

batch_size = 30

# This the maximum number of words that can be passed to the encoder at a time. If a sentence is less than this length, it will be padded with dummy tokens
max_sequence_length = 200

# This is the number of neurons in the feed forward layer
ffn_hidden = 2048

# This is the number of encoder layer that will be repeated
num_layers = 5
```

```python
encoder = Encoder(d_model, ffn_hidden, num_heads, drop_prob, num_layers)
x = torch.randn((batch_size, max_sequence_length, d_model))
out = encoder(x)
```
